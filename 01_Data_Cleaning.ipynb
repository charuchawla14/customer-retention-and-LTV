```python
import pandas as pd 

df = pd.read_csv("online_retail_II.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
# 1. how many rows have missing Customer ID
print(df['Customer ID'].isnull().sum())
```

```python
# 2. Dropping rows with missing Customer ID 
df = df.dropna(subset=['Customer ID'])
```

```python
# 3. Checking for cancelled orders
cancelled = df[df['Invoice'].astype(str).str.startswith('C')]
print(f"Cancelled orders: {len(cancelled)}")
```

```python
# 4. Removing cancelled orders 
df = df[~df['Invoice'].astype(str).str.startswith('C')]
```

```python
# 5. Removing rows with negative or zero Quantity/Price
df = df[(df['Quantity'] > 0) & (df['Price'] > 0)]
```

```python
# 6. Converting InvoiceDate to actual datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
# 7. Converting Customer ID to integer 
df['Customer ID'] = df['Customer ID'].astype(int)
```

```python
# 8. Create TotalPrice column (needed for Monetary in RFM)
df['TotalPrice'] = df['Quantity'] * df['Price']
```

```python
print(df.shape)
df.info()
```

```python
print(f"Original rows: 1,067,371")
print(f"Missing Customer ID removed: {243007}")  # from earlier
print(f"Cancelled orders + negative/zero values removed: {1067371 - 243007 - 805549}")
print(f"Final clean rows: {805549}")
print(f"% of original data retained: {805549/1067371*100:.1f}%")
```

```python
df.to_csv("cleaned_data.csv", index=False)

print("Cleaned data saved successfully!")
```
