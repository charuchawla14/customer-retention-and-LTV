```python
import pandas as pd

rfm = pd.read_csv("rfm_results.csv")
```

```python
rfm.head()
```

```python
from scipy import stats

at_risk = rfm[rfm['Customer_Segment'] == 'At Risk']['Monetary']
regular = rfm[rfm['Customer_Segment'] == 'Regular/Needs Attention']['Monetary']

t_stat, p_value = stats.ttest_ind(at_risk, regular, equal_var=False)

print(f"At Risk mean: £{at_risk.mean():.2f}")
print(f"Regular mean: £{regular.mean():.2f}")
print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.6f}")
```
